# Deaths (Race & Sex) by Covid19 in Georgia

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('deaths.csv')
```

## Deaths by Race

```python
data = dict()
race = df['race'].unique()
for i in range(0, len(race)):
    data[race[i]] = 0

white = df.groupby('race').size()['White']
black = df.groupby('race').size()['African-American/ Black']
asian = df.groupby('race').size()['Asian']
indian = df.groupby('race').size()['American Indian/ Alaska Native']
hawaiian = df.groupby('race').size()['Native Hawaiian/ Pacific Islander']
other = df.groupby('race').size()['Other']
unknown = df.groupby('race').size()['Unknown']

data['White'] = white
data['African-American/ Black'] = black
data['Asian'] = asian
data['American Indian/ Alaska Native'] = indian
data['Native Hawaiian/ Pacific Islander'] = hawaiian
data['Other'] = other
data['Unknown'] = unknown
counties = list(data.keys())
values = list(data.values())
plt.barh(counties, values)
plt.show()
```

## Deaths by Sex

```python
sex = df['sex'].unique()
data2 = {
    'Male' : 0,
    'Female' : 0
}

male = df.groupby('sex').size()['Male']
female = df.groupby('sex').size()['Female']

data2['Male'] = male
data2['Female'] = female

fig = plt.figure(figsize = (20, 10))
plt.title('Death by Sex')
plt.pie(data2.values(), labels = data2.keys(), autopct = '%.1f%%')
plt.legend(loc = (1.0, 0.5))
plt.show()
```
